```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from utils.text_preprocessing import clean_text

import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 200)

plt.style.use("ggplot")
```

```python
df = pd.read_csv("../dataset/fake_job_postings.csv")

df.head()
```

```python
df_clean = df.copy()
```

```python
df_clean.drop(columns=["job_id"], inplace=True)
```

```python
df_clean.head()
```

```python
df_clean.shape
```

```python
text_columns = [
    "title",
    "company_profile",
    "description",
    "requirements",
    "benefits"
]

df_clean[text_columns] = df_clean[text_columns].fillna("")
```

```python
df_clean[text_columns].isnull().sum()
```

```python
df_clean["combined_text"] = (
    df_clean["title"] + " " +
    df_clean["company_profile"] + " " +
    df_clean["description"] + " " +
    df_clean["requirements"] + " " +
    df_clean["benefits"]
)
```

```python
df_clean[["combined_text"]].head()
```

```python
df_clean["clean_text"] = df_clean["combined_text"].apply(clean_text)
```

```python
print("Before Cleaning:\n")
print(df_clean["combined_text"][0])

print("\n" + "="*80 + "\n")

print("After Cleaning:\n")
print(df_clean["clean_text"][0])
```

```python
df_clean.to_csv(
    "../dataset/processed_fake_job_postings.csv",
    index=False
)
```
